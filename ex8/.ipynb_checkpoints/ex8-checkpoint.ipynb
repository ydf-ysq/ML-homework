###  Gaussian distribution
- mean: 
<img src = "pic/mean.jpg" width= 100>
- variance:
<img src = "pic/var.jpg" width= 150>
- The Gaussian distribution is given by:
<img src = "pic/gaussian.jpg" width= 350>

```python
def estimate_gaussian_parameters(X):
    mean = X.mean(axis=0)
    variance = X.var(axis=0)
    return mean, variance
```

```python
def estimate_gaussian(mean, variance, X):
    p = 1 / (np.power(2 * np.pi * variance, 1/2)) * np.exp(-(np.power(X - mean, 2))/(2 * variance))
    p = np.prod(p, axis=1)
    return p
```

### visualizing the contours

```python
def plot_countour(mean, variance, X):
    plt.figure(figsize=(12, 8))
    x = np.linspace(0, 30, 100)
    y = np.linspace(0, 30, 100)
    xx, yy = np.meshgrid(x, y) 
    zz = np.c_[xx.ravel(), yy.ravel()]
    p = estimate_gaussian(mean, variance, zz) 
    p = p.reshape(xx.shape) 
    cont_levels = [10 ** h for h in range(-20, 0, 3)]
    plt.contour(xx, yy, p, cont_levels)
    plt.scatter(X[:, 0], X[:, 1])
```

```python
mean, variance = estimate_gaussian_parameters(X)
plot_countour(mean, variance, X)
plt.show()
```

### Selecting the threshold, ε
在进行高斯分布算法之后会得到矩阵p(x)，里面存放着数据集集每行数据的概率，要判断是否是异常点，我们需要找出一个参数ε，以此作为分界点，比当前ε小的概率p就会被视为异常点，反之为正常点，但是这个参数ε需要不停的尝试才行，所以需要一个评分函数来对每一个ε进行评分，来取最适合的分界点ε。  
在练习的这一部分中，你将使用交叉验证集上的F1分数来实现一种选择分界点ε的算法。
- F1 score：
<img src = "pic/F1.jpg" width = 150>
- precision:
<img src = "pic/prec.jpg" width = 150>
- recall: 
<img src = "pic/rec.jpg" width = 150>

```python
# 使用训练集的平均值和方差
def select_threshold(mean, variance, Xval, yval):
    pval = estimate_gaussian(mean, variance, Xval)
    epsilon = np.linspace(np.min(pval), np.max(pval), num=1000)
    
    f_score = []
    for e in epsilon:
        y_pred = (pval <= e).astype('int')
        f_score.append(f1_score(yval, y_pred))
    
    argmax_fs = np.argmax(f_score)
    
    return epsilon[argmax_fs], f_score[argmax_fs] 
```

```python
e, f_score = select_threshold(mean, variance, Xval, yval)
print('Best epsilon: {}\nBest F-score on validation data: {}'.format(e, f_score))
```

```python
# 找出异常点并在图像上标出
p = estimate_gaussian(mean, variance, X)
weird_point = np.where(p < e)[0]
plot_countour(mean, variance, X)
plt.scatter(X[weird_point,0],X[weird_point,1],edgecolors='r',s=300,facecolors='none')
plt.show()
```

## 2. Recommender Systems
此部分目的是给新用户推荐10个新电影，利用以往的用户对电影的评分，通过协同过滤算法来求出X与Theta矩阵，ex8movies.mat的数据集中存放着Y矩阵（1682行*943列数据），行列分别代表了1682篇电影与943个用户，矩阵中存放的是用户对电影的打分。其中还有R矩阵（1682，943），矩阵中只有0和1俩个值，0代表用户没有看过此电影，1代表看过并且打分.

```python
data = loadmat('data/ex8_movies.mat')
Y = data['Y']  # 不用用户对于不同电影的评分，行为电影数，列为用户数（维度：1682，943）
R = data['R']  # 判断是否有评分，维度与Y矩阵相同
```

```python
data2 = loadmat('data/ex8_movieParams.mat')
X = data2['X']  # 电影类型偏向的特征值，（维度：1682，10）
Theta = data2['Theta']  # 用户的特征值，（维度：943，10）
num_users = int(data2['num_users'])  # 用户数量，943
num_movies = int(data2['num_movies'])  # 电影数量，1682
num_features = int(data2['num_features'])  # 特征数量(X和Theta矩阵的)， 10
```

```python
def serialize(X, Theta):
    return np.r_[X.ravel(), Theta.ravel()] 

def deserialize(param, num_movies, num_users, num_features):
    X = param[:num_movies * num_features].reshape(num_movies, num_features)
    Theta = param[num_movies * num_features:].reshape(num_users, num_features)
    return X, Theta
```

### Collaborative filtering learning algorithm
- cost function:
<img src = "pic/cost.jpg" width = 470>
- gradient:
<img src = "pic/grad.jpg" width = 280>

```python
def cofi_cost_function(param, num_movies, num_users, num_features, Y, R, Lambda):
    X, Theta = deserialize(param, num_movies, num_users, num_features)
    variance = np.power((X @ Theta.T - Y) * R, 2).sum() # X1@Theta.T-Y为所有预测的值偏差，与矩阵R进行点成，筛选出看过的电影
    reg1 = Lambda * np.power(Theta, 2).sum()
    reg2 = Lambda * np.power(X, 2).sum()
    return (variance + reg1 + reg2) * (1 / 2)
```

```python
def gradient(param, num_movies, num_users, num_features, Y, R, Lambda):
    X, Theta = deserialize(param, num_movies, num_users, num_features)
    inner = (X @ Theta.T - Y) * R
    X_derivative = inner @ Theta + Lambda * X
    Theta_derivative = inner.T @ X + Lambda * Theta
    return serialize(X_derivative, Theta_derivative)
```

### Learning movie recommendations
输入你自己的电影偏好，以便稍后算法运行时，你可以获得自己的电影推荐。我们已经根据自己的喜好填写了一些值，但您应该根据自己的喜好进行更改。

```python
# 添加新的用户
my_ratings = np.zeros((num_movies, 1))
my_ratings[0] = 5
my_ratings[7] = 5
my_ratings[14] = 4
my_ratings[21] = 4
my_ratings[28] = 3
my_ratings[35] = 3
my_ratings[75] = 2
my_ratings[163] = 2
my_ratings[260] = 1
my_ratings[370] = 5
my_ratings[580] = 5
my_ratings[600] = 5
my_ratings[800] = 5
 
Y = np.c_[Y, my_ratings]
R = np.c_[R, my_ratings != 0]
Y.shape,R.shape
```

```python
# 均值归一化
def mean_formulize(Y, R):
    Y_mean = Y.sum(1) / R.sum(1)
    Y_formu = (Y - Y_mean.reshape(Y.shape[0], 1)) * R
    return Y_formu,Y_mean
```

```python
Y_formu,Y_mean = mean_formulize(Y, R)
num_movies, num_users = Y_formu.shape
num_features = 10
X = np.random.random((num_movies, num_features))
Theta = np.random.random((num_users, num_features))
params = serialize(X, Theta)
Lambda = 10

res = minimize(fun=cofi_cost_function, 
               x0=params, 
               args=(num_movies, num_users, num_features, Y_formu, R, Lambda),
               method='TNC', jac=gradient, options={'maxfun': 100})
X_res, Theta_res = deserialize(res.x, num_movies, num_users, num_features)
print(X_res.shape, Theta_res.shape)
```

```python
# 下面开始预测
prediction_matrix = X_res @ Theta_res.T + Y_mean.reshape(num_movies,1)
last_user = prediction_matrix[:,-1]
sort_list = np.argsort(-last_user)
top_ten = sort_list[:10]

# 导入电影名字
movies = []  # 包含所有电影的列表
with open('data/movie_ids.txt','r', encoding='latin 1') as f:
    for line in f:
        movies.append(' '.join(line.strip().split(' ')[1:]))

# 取出推荐的10部电影
for i in top_ten:
    print('Predicting rating {} for movie {}'.format(last_user[i], movies[i])) 
```
